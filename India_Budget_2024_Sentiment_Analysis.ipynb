```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.pib.gov.in/PressReleasePage.aspx?PRID=1895315#:~:text=Budget%20Estimates%202023%2D24%3A,5.9%20per%20cent%20of%20GDP.'
response = requests.get(url)

scraped_data=[]

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    list_items = soup.find_all('li')

    if list_items:
        for item in list_items:
            text = item.get_text().strip()
            if len(text)!=0:
                print(text)
                scraped_data.append(text)
    else:
        print("No <li> items found on the webpage.")
else:
    print("Failed to retrieve the webpage")
```

```python
from transformers import pipeline

pipe = pipeline("sentiment-analysis",model="siebert/sentiment-roberta-large-english")
```
